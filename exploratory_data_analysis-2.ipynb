```python
#showing the numerical and categorical features
def type_features(data):
    numerical_features = data.select_dtypes(exclude = ["object"]).columns
    print("numerical_features:",numerical_features)
    categorical_features = data.select_dtypes(include = ["object"]).columns
    print( "categorical_features :",categorical_features)
    

type_features(tourism_df)
```

```python
#histogram of age distribuition
plt.figure(figsize=(8, 5))
sns.histplot(tourism_df["Age"], bins=10,kde=True, color="black")

# Labels and title
plt.xlabel("Age")
plt.ylabel("Number of Tourists")
plt.title("Age Distribution of Tourists")

# Show the plot
plt.show()
```

```python
#boxplot for tourist ratings
#boxplot for tourist satisfaction 
plt.figure(figsize=(6,5))
sns.boxplot(x=tourism_df["Tourist Rating"], color="yellow")
plt.xlabel('tourist rating')
plt.title('Boxplot of tourist rating', fontsize=15 )
plt.show()
plt.figure(figsize=(6,5))
sns.boxplot(x=tourism_df["Satisfaction"], color="green")
plt.xlabel('satisfaction')
plt.title('Boxplot of tourist satisfaction', fontsize=15)
plt.show()
```

```python
#Line Chart of System response time vs. tourist rating 

plt.figure(figsize=(8, 5))
sns.lineplot(x=tourism_df["System Response Time"], y=tourism_df["Tourist Rating"], marker="o", color="blue")


plt.xlabel("System Response Time (seconds)")
plt.ylabel("Tourist Rating")
plt.title("System Response Time vs. Tourist Rating")
plt.show()
```

```python
#Bar Chart of Average tourist rating for different sites.
avg_ratings=tourism_df.groupby("Site Name")["Tourist Rating"].mean().sort_values()
plt.figure(figsize=(10,5))
sns.barplot(x=avg_ratings.index, y= avg_ratings.values, palette="viridis")
plt.xlabel("Site Name")
plt.ylabel("Average Tourist Rating")
plt.title("Average Tourist Rating for Different Sites")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
#plot for satisfaction level grouped by accessibility
satisfaction_accessibility=tourism_df.groupby(["Satisfaction", "Accessibility"]).size().unstack()
satisfaction_accessibility.plot(kind="bar", figsize=(8, 5), colormap="viridis")
plt.xlabel("Accessibility (True/False)")
plt.ylabel("Number of Tourists")
plt.title("Satisfaction Levels Grouped by Accessibility")
plt.legend(title="Satisfaction Level")
plt.xticks(rotation=0)
plt.show()
```

this graph shows that there is inverse relation between accessibility and number od tourists

```python
#plot for interest vs tourist count
from collections import Counter
interests_list = tourism_df["Interests"].apply(eval).explode()
# Count occurrences of each interest
interest_counts = Counter(interests_list)

# Convert to DataFrame
interest_df = pd.DataFrame(interest_counts.items(), columns=["Interest", "Tourist Count"]).sort_values(by="Tourist Count", ascending=False)

# Plot Stacked Bar Chart
plt.figure(figsize=(10, 5))
sns.barplot(x=interest_df["Interest"], y=interest_df["Tourist Count"], palette="magma")

# Labels and title
plt.xlabel("Interest")
plt.ylabel("Number of Tourists")
plt.title("Tourist Count by Interest")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
#scatter plot for tourist rating vs VR experience quality 
# Plot Scatter Plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x=tourism_df["Tourist Rating"], y=tourism_df["VR Experience Quality"], color="green")

# Labels and title
plt.xlabel("Tourist Rating")
plt.ylabel("VR Experience Quality")
plt.title("Tourist Rating vs. VR Experience Quality")

# Show the plot
plt.show()
```

```python
#heatmap for Correlation Between Numerical Attributes
numerical_cols = ["Recommendation Accuracy", "System Response Time", "Satisfaction"]
correlation_matrix = tourism_df[numerical_cols].corr()

# Plot Heatmap
plt.figure(figsize=(7, 5))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Title
plt.title("Correlation Heatmap of Numerical Attributes")

# Show the plot
plt.show()
```

```python
# Pie Chart for Distribution of Interests
# Flatten the list of interests (converting string representation of lists into actual lists)
interests_list = tourism_df["Interests"].apply(eval).explode()

# Count occurrences of each interest
interest_counts = Counter(interests_list)

# Convert to DataFrame
interest_df = pd.DataFrame(interest_counts.items(), columns=["Interest", "Count"])
plt.figure(figsize=(8, 8))
plt.pie(interest_df["Count"], labels=interest_df["Interest"], autopct="%1.1f%%", colors=sns.color_palette("pastel"))
plt.title("Distribution of Tourist Interests")
plt.show()
```

this pie chart shows that the highest number of people are interested in nature, or art and culture
